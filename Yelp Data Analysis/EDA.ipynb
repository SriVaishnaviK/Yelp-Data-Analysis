```
from pyspark.sql.functions import hour, dayofweek
import matplotlib.pyplot as plt
import seaborn as sns

merged_df_check = merged_df_check.withColumn("checkin_hour", hour("date"))
merged_df_check = merged_df_check.withColumn("checkin_dayofweek", dayofweek("date"))
checkin_by_hour = merged_df_check.groupBy("checkin_hour").count().orderBy("checkin_hour")

checkin_by_dayofweek = merged_df_check.groupBy("checkin_dayofweek").count().orderBy("checkin_dayofweek")

sns.lineplot(data=checkin_by_hour.toPandas(), x="checkin_hour", y="count")
plt.xlabel("Hour of the Day")
plt.ylabel("Check-in Count")
plt.title("Check-in Patterns by Hour")
plt.show()


sns.barplot(data=checkin_by_dayofweek.toPandas(), x="checkin_dayofweek", y="count")
plt.xlabel("Day of the Week")
plt.ylabel("Check-in Count")
plt.title("Check-in Patterns by Day of the Week")
plt.show()
```

```
from pyspark.sql.functions import month
import matplotlib.pyplot as plt
import seaborn as sns
checkin_df = checkin_df.withColumn("checkin_month", month("date"))

checkin_by_month = checkin_df.groupBy("checkin_month").count().orderBy("checkin_month")

sns.lineplot(data=checkin_by_month.toPandas(), x="checkin_month", y="count")
plt.xlabel("Month")
plt.ylabel("Check-in Count")
plt.title("Seasonal Trends in Check-in Data")
plt.show()
```
