### Trend of Ratings since the beginning

```
def calculate_scaled_avg_rating(df):
    # group by year and business_id, and calculate the average rating and count
    grouped_df = df.groupby(year('date').alias('year')) \
                             .agg(avg('stars').alias('avg_stars'), count('*').alias('count'))

    year_count = grouped_df.groupby('year').agg(sum('count').alias('total_count'))
    joined_df = grouped_df.join(year_count, on='year')
    scaled_df = joined_df.withColumn('scaled_avg', round((col('avg_stars') * col('count')) / (col('total_count')), 2))
    result_df = scaled_df.select('year', 'scaled_avg', 'total_count').orderBy('year')
    return result_df
```

```
lowest_b_rating_trend = calculate_scaled_avg_rating(lowest_rated_df)
highest_b_rating_trend = calculate_scaled_avg_rating(highest_rated_df)
avg_b_rating_trend = calculate_scaled_avg_rating(average_rated_df)
```

### Sentiment Analysis of Customer Review per year.

```
from pyspark.sql.functions import col, year, month, explode

def filter_reviews_by_year(sample_year, search_words, df):
    # Filter reviews for the sample year          
    filtered_df = df.filter(year('date') == sample_year)
    # Select relevant columns and explode the 'noun_sentiment' map column
    filtered_df = filtered_df.select("business_id", "stars", explode("noun_sentiment").alias("noun", "sentiment"))

    # Filter the reviews based on the search words
    filtered_df = filtered_df.filter(col('noun').isin(search_words))
    
    return filtered_df
```

```
# Reviews for Lowest rated restaurant
lowest_review_df = filter_reviews_by_year(2016, search_words, lowest_rated_df)
# Reviews for 
best_review_df = filter_reviews_by_year(2019, search_words, highest_rated_df)
avg_review_df = filter_reviews_by_year(2019, search_words, average_rated_df)
```
